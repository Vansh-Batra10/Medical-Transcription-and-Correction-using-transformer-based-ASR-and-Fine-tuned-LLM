### Import required packages

```python
import torch
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor, pipeline
```

```python
device = "cuda:0" if torch.cuda.is_available() else "cpu"
torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32
```

```python
device
```

```python
torch_dtype
```

```python
model_id = "openai/whisper-large-v3"
```

```python
model = AutoModelForSpeechSeq2Seq.from_pretrained(
    model_id, torch_dtype=torch_dtype, low_cpu_mem_usage=True, use_safetensors=True
)
model.to(device)

processor = AutoProcessor.from_pretrained(model_id)
```

```python
pipe = pipeline(
    "automatic-speech-recognition",
    model=model,
    tokenizer=processor.tokenizer,
    feature_extractor=processor.feature_extractor,
    max_new_tokens=128,
    chunk_length_s=15,
    batch_size=16,
    return_timestamps=True,
    torch_dtype=torch_dtype,
    device=device,
)
```

```python
pipe
```

```python
audio = ""
```

```python
result = pipe(audio) # v2 gives better results if you don't provide a language.
```

```python
import pytube
from pydub import AudioSegment
import tempfile
download_directory = tempfile.mkdtemp()
def download_youtube_video(video_url, download_dir):
    try:
        yt = pytube.YouTube(video_url)
        stream = yt.streams.filter(only_audio=True).first()
        video_file_path ='video_audio.mp4'
        stream.download(output_path="./video", filename='video_audio')
        return video_file_path
    except Exception as e:
        print(f"Error downloading video: {str(e)}")
        return None

def convert_audio_to_wav(video_file_path, download_dir):
    audio_file_path = './video/video_audio'

    audio = AudioSegment.from_file(audio_file_path, format="mp4")
    audio.export(audio_file_path, format="mp3")

    return audio_file_path

def transcribe_audio(audio_file_path):
    # Use your speech-to-text model or API here
    audio="./video/video_audio"
    result = pipe(audio)
    return result.get('text', '')

def transcribe_video_audio(video_url, download_dir):
    video_file_path = download_youtube_video(video_url, download_dir)

    if video_file_path is not None:
        audio_file_path = convert_audio_to_wav(video_file_path, download_dir)
        transcribed_text = transcribe_audio(audio_file_path)
        return transcribed_text

transcribed_texts = []


# Assuming video_links is a list of YouTube video URLs
for video_url in video_links:
    transcribed_text = transcribe_video_audio(video_url, download_directory)
    if transcribed_text:
        transcribed_texts.append(transcribed_text)

# Print the transcribed text for each video
for i, text in enumerate(transcribed_texts, start=1):
    print(f"Transcribed Text for Video {i}:")
    print(text)
```
